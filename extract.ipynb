```python
import pandas as pd, matplotlib as plt, numpy as np
dataset = pd.read_excel("coffee_sales.xlsx")
dataset['date'] = pd.to_datetime(dataset['date'])
dataset = dataset[dataset['date'].between('2024-03-01', '2025-02-28')]
dataset
```

```python
#tabela de total por café 

by_coffee = dataset.groupby('coffee_name')['money'].sum().sort_values(ascending=False).reset_index()
by_coffee.to_csv('by_coffee.csv')
```

```python
#tabela de total por mês
dataset['mes_ano'] = dataset['date'].dt.to_period('M')
by_month = dataset.groupby('mes_ano')['money'].sum().reset_index()
by_month['mes_ano_text'] = by_month['mes_ano'].dt.strftime('%b %Y')
by_month = by_month.drop(columns=['mes_ano'])
order = ['mes_ano_text', 'money']
by_month = by_month[order]
by_month.to_csv('by_month.csv')
```
